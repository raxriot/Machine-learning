```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/digipodium/Datasets/main/regression/Salary_Data.csv')
df.head()
```

```python
plt.scatter(data=df, x = 'YearsExperience', y ='Salary', marker = 'o', color = 'red', alpha=0.5)
plt.show()
```

```python
x, y = df[['YearsExperience']], df['Salary']
model =LinearRegression()
#Train the model
model.fit(x, y)
```

```python
#evaluation
score = model.score(x, y)
print('accuracy', score*100)
```

```python
#prediction 
ypred = model.predict(x)
print(ypred)
```

```python
#vizuLization
plt.scatter(x,y, marker = 'o', color = 'red', alpha=0.5)
plt.plot(x, ypred, color ='green', linewidth=5, alpha=0.5)
plt.title('Salary vs Experience (regression)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
df = pd.read_csv('../datasets/data.csv')
df
```

```python
plt.scatter(data=df, x = 'x', y ='y', marker = 'o', color = 'red', alpha=0.5)
plt.show()
```

```python
#split data
x, y = df [['x']], df ['y']
Xtrain, Xtest, ytrain, ytest = train_test_split(x, y)
```

```python
# train
model = LinearRegression()
model.fit(Xtrain, ytrain)
```

```python
# evaluation
ypred = model.predict(Xtest)
accuracy = r2_score(ytest, ypred)#original, predicted
mse = mean_squared_error(ytest, ypred)
mae= mean_absolute_error(ytest, ypred)
print("accuracy", accuracy*100)
print('mse',mse)
print('mae',mae)
```

```python
#vizuLization
plt.scatter(x,y, marker = 'o', color = 'red', alpha=0.5, label = 'original')
plt.plot( x, model.predict(x), color ='green', linewidth=5, alpha=0.5, label = 'predicted')
plt.title('Linear Regression(full data)')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend
plt.show()
 
```

```python
from joblib import dump, load
import os
```

```python
os.makedirs('models', exist_ok=True)# create a folder if not exists
#save model
with open ('models/linreg_model1.jb','wb') as f:
    dump(model,f)
```
